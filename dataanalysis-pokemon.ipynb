````python
#cara lain mengetahui nilai NaN dalam df di setiap kolom
poke.isna().sum()
````

Ketika yakin bahwa baris NA dapat dihapus dengan aman maka dapat digunakan fungsi **dropna()** dan dapat diatur threeshold, secara default menghapus bari jika ada nilai NA (`how='any'`). tetapi metode ini dapat diatur jika semua nilai adalah NA dari baris itu (`how='all'`)

```
    # drops row if all values are NA
    poke.dropna(how='all')
    
    # drops row if it doesn't have at least 5 non-NA values
    poke.dropna(thresh=5) 
```

````python
##buat copy dari file sebelumnya agar bisa dibandingkan untuk lebih baik dihapus atau difill
poke2 = poke.copy()
poke2.head()
````

````python
poke2['Type 2'] = poke2['Type 2'].fillna('Missing')

poke2.head() #hasil pada baris ke 4 Nan menjadi Missing
````

## Duplicated Data

Untuk mengamati duplikat dalam data kita, kita dapat menggunakan `duplicate()` dan menggabungkannya dengan metode subsetting seperti di bawah ini:

````python
poke2[poke2.duplicated(keep=False)] #tidak ada data duplikat
````

## Sorting Data

````python
poke.sort_values(['Type 1', 'Attack'], ascending=[True,True])
````

## Adding a Total collumn

````python
#poke['Total'] = poke['HP'] + poke['Attack'] + poke['Defense'] + poke['Sp. Atk'] + poke['Sp. Def'] + poke['Speed']

poke2['Total'] = poke.iloc[:,4:10].sum(axis=1)
poke2.head()
````

## Filtering Data


````python
#menggunakan operator and dengan 3 kondisi
poke2.loc[(poke2['Type 1'] == 'Fire') & (poke2['Type 2'] == 'Rock') & (poke2['Total']>300)]
````

````python
#mencari data dengan str.contains

poke2.loc[poke2['Name'].str.contains('saur')] 
````

````python
#tanpa mengandung huruf saur di kolom nama dengan tambahan ~
poke2.loc[~poke2['Name'].str.contains('saur')] 
````

````python
#menggunakan regex untuk filtering
import re

poke2.loc[poke2['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
````

## Aggregate Data

````python
#mencari tau type dengan defense terbagus berdasar rata-rata
poke2.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
````

## Saving Data

````python
poke2.to_csv('pokemon_mod.csv', index=False)
````
